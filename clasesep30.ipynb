```python
# np.arrange(start, end, step)
x = np.arange(0,3*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
ysin = np.sin(x)
ycos = np.cos(x)
plt.plot(x,ysin)
plt.plot(x,ycos)

plt.xlabel('X axis label')
plt.ylabel('Y axis label')

plt.title(' sine & cosine functions')
plt.show()
```

```python
plt.subplot(2,1,1) # 2 rows, 1 column, position 1

plt.plot(x,ysin)
plt.title('Sine')

plt.subplot(2,1,2) # 2 rows, 1 column, position 2
plt.plot(x,ycos)
plt.title('Cosine')

# para 4 en 1 plt.subplot(2,2,1-4)

plt.show()
```

```python
import seaborn as sns
```

```python
path = 'datasets/Cartwheeldata.csv'
df = pd.read_csv(path)
df.head(2)
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'Height',
           data = df,
           #fit_reg = False , para quitar linea de regresion
           hue = 'Gender'
          )
plt.title('Scatter plot of Wingspan vs Height')
plt.show()
```

```python
sns.swarmplot(x='Gender',
              y='Wingspan',
              data =df,
             )

plt.title('Gender vs Wingspan, the Swarmplot')
plt.show()
```

```python
sns.boxplot(data = df.loc[df['Gender']=='M',['Age','Height','Wingspan', 'CWDistance' ]])

plt.title('Male boxplot')
```

```python
sns.boxplot(data = df.loc[df['Gender']=='F',['Age','Height','Wingspan', 'CWDistance' ]])
plt.title('Female boxplot')
```

```python
sns.distplot(df.CWDistance)
```

```python
sns.countplot(x='Gender',
              data =df)
plt.xticks(rotation =45)
plt.yticks(rotation =45)
plt.show()
```
